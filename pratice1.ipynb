#### Derived Columns - Vectorise not iterrows

```python
# Accelerometer magnitude
df['acc_magnitude'] = np.sqrt(df['acc_x']**2 + df['acc_y']**2 + df['acc_z']**2)
print(df['acc_magnitude'])
```

```python
# Gyro magnitude
df['gyro_magnitude'] = np.sqrt(df['gyro_x']**2 + df['gyro_y']**2 + df['gyro_z']**2)
print(df['gyro_magnitude'])
```

```python
# Boolean flag column
df['is_clipped'] = df['acc_z'].abs() > 16.0 # sensor saturation threshold
# print(df['is_clipped'])
print(df['is_clipped'].value_counts()) # check total values
print(df['is_clipped'].sum()) # get only true value counts
```

```python
# np.where - conditional column
df['motion_class'] = np.where(df['acc_magnitude'] > 2.0, 'high_motion', 'low_motion')
print(df['motion_class'])
print(df['motion_class'].value_counts())
```

```python
# why not iterrows
for idx, row in df.iterrows():
    df.loc[idx, 'mag'] = (row.acc_x**2 + row.acc_y**2 + row.acc_z**2)**0.5
print(df['mag'])

# same result
df['acc_magnitude'] = np.sqrt(df['acc_x']**2 + df['acc_y']**2 + df['acc_z']**2)
print(df['acc_magnitude'])
```

#### GroupBy & Aggregation

```python
# single aggregation
df.groupby('activity')['acc_magnitude'].mean()
```

```python
# multiple aggregations
summary = df.groupby('activity').agg(
    n_rows = ('acc_x', 'count'),
    acc_mean = ('acc_magnitude', 'mean'),
    acc_std = ('acc_magnitude', 'std'),
    gyro_mean = ('gyro_magnitude', 'mean'),
).reset_index()
print(summary)
```

```python
# per participant session count
df.groupby('participant_id')['session_id'].nunique().reset_index(name='n_sessions')
```

```python
# class distribution per participant - normailised
dist = df.groupby(['participant_id', 'activity']).size().unstack(fill_value=0)
print(dist)
dist_pct = dist.div(dist.sum(axis=1), axis=0).round(3)
print(dist_pct)
```

#### Time Series Operations - Critical for Sensor Data

```python
# Parse timestamps if not already datetime
df['timestamp'] = pd.to_datetime(df['timestamp'])
print(df['timestamp'])
df = df.sort_values('timestamp').reset_index(drop=True)
print(df)
```

```python
# check monotonicity - non monotonicity = sensor bug/ data corruption
is_monotonic = df['timestamp'].is_monotonic_increasing
print(is_monotonic)
if not is_monotonic:
    bad_rows = df[df['timestamp'] < df['timestamp'].shift(1)]
    print(f'Non-monotonic timestamps at rows: {bad_rows.index.tolist()}')
```

```python
# timestamps diffs - detect gaps
df['ts_diff_ms'] =  df['timestamp'].diff().dt.total_seconds() * 1000
print(df['ts_diff_ms'])
expected_ms = 10 # 100Hz = 10ms between samples
gaps = df[df['ts_diff_ms'] > expected_ms * 3]
print(gaps)
```

```python
# rolling stats (window stats)
df['acc_mag_rolling_mean'] = df['acc_magnitude'].rolling(window=50, min_periods = 1).mean()
print(df['acc_mag_rolling_mean'])
df['acc_mag_rolling_std'] = df['acc_magnitude'].rolling(window=50, min_periods = 1).std()
print(df['acc_mag_rolling_std'])
```

```python
# resample (reduce sampling rate)
df_indexed = df.set_index('timestamp')
df_25hz = df_indexed[['acc_x', 'acc_y', 'acc_z']].resample('40ms').mean()
print(df_25hz)
```

#### Merging & Joining

```python
# Left join - keep all rows from seesions, add metadata where available
# print(df.head(5))
df_sessions = df[['session_id']] 
df_metadata = df[['session_id', 'timestamp', 'acc_x', 'acc_y', 'acc_z', 'participant_id', 'activity']]
merged = pd.merge(df_sessions, df_metadata, on='session_id', how='left')
print(merged.head(5))
```

```python
# Check for merge failures (right side had no match)
print(merged.columns.tolist())
unmatched = merged[merged['participant_id'].isna()]
print(f'{len(unmatched)} sessions with no metadata match')
```

```python
# Concat multiple dataframes (eg combining all session CSVs)
import glob
all_session_dfs = []
for path in glob.glob('./*.csv'):
    tmp = pd.read_csv(path)
    tmp['source_file'] = path
    all_session_dfs.append(tmp)
combined = pd.concat(all_session_dfs, ignore_index=True)
print(combined)
```

#### Handling Missing Values

```python
# Null rate per column
null_rates = df.isnull().mean()
print(null_rates)
problematic = null_rates[null_rates != 0]
print(problematic)
```

```python
# Drop rows with any null in key sensor columns
df_clean = df.dropna(subset=['acc_x', 'acc_y', 'acc_z'])
df_clean.shape
```

```python
# forward fill (carry last known value) - OK for slow changing signals
df['activity'] = df['activity'].ffill()
print(df.isnull().mean())
```
