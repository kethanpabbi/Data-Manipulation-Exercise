### Sorting

#### 1. Bubble Sort

##### Repeatedly step through the list, compare adjacent elements, and swap them if they are in the wrong order. After each pass the largest unseen element 'bubbles' to its correct position at the end. O(n²) — never use in production. Mainly asked to test if you understand comparison-based sorting.

```python
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        swapped = False
        for j in range(0, n-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
                swapped = True
        if not swapped:
            break
    return arr

print(bubble_sort([5,6,1,4,9,8,7,2,10]))
```

#### 2. Selection Sort

##### Find the minimum element in the unsorted portion and swap it to the front. Unlike bubble sort which does many small swaps, selection sort does exactly n-1 swaps total. Still O(n²) but fewer writes — relevant when writes are expensive (e.g. flash memory).

```python
def selection_sort(arr):
    n = len(arr)
    for i in range(n):
        min_idx = i
        for j in range (i+1, n):
            if arr[j] < arr[min_idx]:
                min_idx = j
        arr[i], arr[min_idx] = arr[min_idx], arr[i]
    return arr

print(selection_sort([5,6,1,4,9,8,7,2,10]))
```

#### 3. Insertion Sort

##### Insertion sort builds a sorted sublist by taking one element at a time and inserting it in the correct position. O(n) best case on nearly-sorted data. Python's built-in Timsort uses insertion sort for small subarrays (< 64 elements) because it has low overhead and is cache-friendly. Good answer for interview: 'When n is small, or data is nearly sorted — otherwise merge sort.'

```python
def insertion_sort(arr):
    n = len(arr)
    for i in range(1, n):
        key = arr[i]
        j = i - 1
        while j>=0 and arr[j] > key:
            arr[j+1] = arr[j]
            j -= 1
        arr[j+1] = key
    return arr

print(insertion_sort([5,6,1,4,9,8,7,2,10]))
```

#### 4. Merge Sort

##### Divide the list in half recursively until subarrays of size 1, then merge them back in sorted order. O(n log n) time always — no worst case degradation. O(n) space — needs auxiliary arrays to merge. Stable sort — preserves relative order of equal elements. This is what Python's sorted() is based on (Timsort is an optimised merge sort).

```python
def merge_sort(arr):
    if len(arr) <= 1:
        return arr
    
    mid = len(arr) // 2
    left = merge_sort(arr[:mid])
    right = merge_sort(arr[mid:])
    return merge(left, right)

def merge(left, right):
    result = []
    i = j = 0
    while i < len(left) and j < len(right):
        if left[i] <= right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1
    result.extend(left[i:])
    result.extend(right[j:])
    return result

print(merge_sort([5,6,1,4,9,8,7,2,10]))
```

#### 5. Quick Sort

##### Pick a pivot, partition elements into less-than and greater-than groups, recurse on each. O(n log n) average. Worst case O(n²) happens when the pivot is always the min or max (already sorted input with first-element pivot). Fix: random pivot or median-of-three. O(log n) space on the call stack (in-place).

```python
import random

def quick_sort(arr):
    if len(arr) < 1:
        return arr
    pivot = random.choice(arr)
    # print(f'pivot: {pivot}')
    less = [x for x in arr if x < pivot]
    # print(f'less: {less}')
    equal = [x for x in arr if x == pivot]
    # print(f'equal: {equal}')
    greater = [x for x in arr if x > pivot]
    # print(f'greated: {greater}')
    # print(f'arr: {arr}')
    return quick_sort(less) + equal + quick_sort(greater)

print(quick_sort([5,6,1,4,9,8,7,2,10]))
```

#### 6. Pythons Build-In Sorting

##### Python uses Timsort — a hybrid of merge sort and insertion sort. O(n log n) always, O(n) best case on nearly-sorted data. Stable. sorted() returns a new list. list.sort() sorts in-place. Use the key= parameter for custom sorting — it is called once per element (efficient).

```python
# Basic Sort
nums = [3, 1, 5, 2, 9, 8, 6, 7]
print(f'Sorted numbers: {sorted(nums)}')
print(f'Sorted numbers: {sorted(nums, reverse=True)}')

# sort by string length
words = ['banana', 'apple', 'kiwi', 'mango']
print(f'Sorted by str len: {sorted(words, key = len)}')

# sort list of dicts by field
sessions = [{'id': 'S2', 'rows': 500}, {'id': 'S1', 'rows': 600}, {'id': 'S3', 'rows': 200}]
sorted_session = sorted(sessions, key=lambda x: x['rows'])
print(f'Sorted Sessions: {sorted_session}')

# multi-key sort: first by activity, then by session id
data = [('running', 'S3'), ('walking', 'S1'), ('running', 'S1')]
print(f'Sorted Data: {sorted(data, key=lambda x: (x[0], x[1]))}')
```

#### 2. String Formatting

##### Q: What is an f-string and why is it preferred over .format() or % formatting?
A: f-strings (Python 3.6+) are evaluated at runtime and are the fastest and most readable option. They embed expressions directly in the string literal.



```python
# example
participant = 'P001'
n_rows = 3000
null_rate = 0.0267

# f-string
print(f'Participant {participant}: {n_rows} rows. null rate of {null_rate:.2%}')

# useful formatting
print(f'{n_rows:,}') # comma separator
print(f'{null_rate:.4f}')
print(f'{null_rate:.2%}')
print(f'{participant:>10}')
print(f'{"="*40}')
```

#### 3. Reverse a string

##### Q: How do you reverse a string in Python? What is the most Pythonic way?
A: Use slicing with step -1. Python strings support slice notation [start:stop:step]. s[::-1] means 'all characters, stepping backwards'.


```python
s = 'sensor'

# pythonic way: slices with -1 step
print(f'Pythonic slicing: {s[::-1]}')

# using reversed() + join
print(f'Using Reversed: {reversed(s)}')
print(f"Using Reversed: {''.join(reversed(s))}")

# manual loop
def reversed_string(s):
    result = ''
    for char in s:
        result = char + result
    return result
print(f'Test Manual Loop: {reversed_string("sensor")}')
```

#### 4. Check if Palindrome

##### Q: Write a function to check if a string is a palindrome.
A: A palindrome reads the same forwards and backwards. Compare the string to its reverse. Handle case and spaces if the question asks for 'real word' palindromes.


```python
def is_palindrome(str):
    str = str.lower()
    return str == str[::-1]

# ignore cases
print(f'Check Palindrome for racecar: {is_palindrome("racecar")}')
```

#### 5. Count Char Frequency

##### Q: Count how many times each character appears in a string. What data structure do you use?
A: Use a dictionary or collections.Counter. Counter is the idiomatic Python answer — it is a dict subclass that counts hashable objects.


```python
from collections import Counter

s = "accelerometer"

# counter - pythonic
freq = Counter(s)
print(f'Frequency: {freq}')
print(f'Most common: {freq.most_common(3)}')

# manual dict
freq_manual = {}
for char in s:
    # print(char,freq_manual.get(char,0))
    freq_manual[char] = freq_manual.get(char,0) + 1
print(f'Manual counter: {freq_manual}')

# default dict approach
from collections import defaultdict
freq_dd = defaultdict(int)
for char in s:
    freq_dd[char] += 1
print(f'Default dict: {freq_dd}')
```

#### 6. Check for Anagrams

##### Q: Write a function to check if two strings are anagrams of each other.
A: Two strings are anagrams if they contain the same characters with the same frequencies. Simplest: sort both and compare. More efficient: compare Counter objects.

```python
from collections import Counter

def are_anagrams(s1, s2):
    return Counter(s1.lower()) == Counter(s2.lower())

# sort and compare
def are_anagrams_sort(s1, s2):
    return sorted(s1.lower()) == sorted(s2.lower())

print(f'Using Counter: {are_anagrams("silent", "Listen")}')
print(f'Using Sort: {are_anagrams_sort("SILENT", "LiSteN")}')
```

#### 7. Find all unique chars

##### Q: Find the first non-repeating character in a string.
A: Build a frequency count, then scan the string once more to find the first character with count 1. O(n) total.

```python
from collections import Counter

def first_unique_char(s):
    freq = Counter(s)
    for i, ch in enumerate(s):
        print(i,ch)
        if freq[ch] == 1:
            return ch, i
    return None, -1

print(f'First Unique Char: {first_unique_char("accelerometer")}')
```

#### 8. String Slicing

##### Q: Explain Python string slicing. What does s[2:8:2] mean?
A: s[start:stop:step] — start is inclusive, stop is exclusive, step is the increment. Negative values work backwards. This is one of Python's most powerful features.

```python
s = 'accelerometer'
print(f'First 5: {s[0:5]}')
print(f'From Index 5: {s[5:]}')
print(f'Last 4 chars: {s[-4:]}')
print(f'Every other char: {s[::2]}')
print(f'Reverse: {s[::-1]}')
print(f'Chars at even indx: {s[2:8:2]}')

# practical use
session_id = 'P001_S03_walking'
participant = session_id[:4]
print(f'Participant name: {participant}')
print(f'Activity: {session_id[9:]}')
```
